```python
#All punctuations are replaced with white spaces

df['Sinhala'] = df['Sinhala'].apply(lambda x:re.sub(r"""
               [,.;@#?!&$""/''*]+  # Accept one or more copies of punctuation
               \ *           # plus zero or more copies of a space,
               """,
               " ",          # and replace it with a single space
               x, flags=re.VERBOSE) )

df['English'] = df['English'].apply(lambda x:re.sub(r"""
               [,.;@#?!&$""/''*]+  # Accept one or more copies of punctuation
               \ *           # plus zero or more copies of a space,
               """,
               " ",          # and replace it with a single space
               x, flags=re.VERBOSE) )
```

# From here onwards cleaned data are organized for train the model

```python
#Converting the corpus into two lists of English and Sinhala words. English words will be converted into Inputs and Sihala words will be
#converted into outputs from the algorithm in future steps.

englishList=[e.split() for e in English]
sinhalaList=[s.split() for s in Sinhala]
```

```python
sinwords=[]
enwords=[]
for i,j in zip(englishList,sinhalaList):
  for k,l in zip(i,j):
    enwords.append(k)
    sinwords.append(l)
```

```python
enletters=[list(i) for i in enwords]
sinletters=[list(i) for i in sinwords]
```

```python
enletters
```

```python
enX=[]
for i in enletters:
  words=[]
  for j in i:
    words.append(english_dic.get(j))
  enX.append(words)

sinY=[]
for i in sinletters:
  words=[]
  for j in i:
    words.append(sinhala_dic.get(j))
  sinY.append(words)
```

```python
encoderInputs = np.ones((19714,22,60))
for i, sample in enumerate(enX):
  for j,character in enumerate(sample):
     for k, value in enumerate(character):
       encoderInputs[i][j][k]=value
```

```python
decoderOutputs= np.ones((19714,22,77))
for i, sample in enumerate(sinY):
  for j, character in enumerate(sample):
    for k, value in enumerate(character):
      decoderOutputs[i][j][k]=value   
```

```python
encoderInputs
```

# Bidirectional LSTM Model

```python
X_train, X_test, y_train, y_test = train_test_split(encoderInputs, decoderOutputs, test_size=0.25)
```

```python
inference_dic = {} 
for key in sinhala_dic:
   inference_dic[key] = np.argmax(sinhala_dic[key])

dic_inference = dict((v,k) for k,v in inference_dic.items())
       
```

```python
inference_dic_pickle=open("/content/drive/MyDrive/Project/inference_dic.pickle","wb")
pickle.dump(dic_inference,inference_dic_pickle)
inference_dic_pickle.close()
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(128, input_shape=(22,60), return_sequences=True)))
model.add(Bidirectional(LSTM(128,return_sequences=True)))
#model.add(Bidirectional(LSTM(256,return_sequences=True)))
model.add(TimeDistributed(Dense(128,activation="relu")))
model.add(TimeDistributed(Dense(77,activation="softmax")))
model.compile(loss="binary_crossentropy", optimizer='adam' , metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, validation_split=0.20, epochs=100, batch_size=20, verbose=2)
```

```python
inf_english_dic = {} 
for key in english_dic:
   inf_english_dic[key] = np.argmax(english_dic[key])
```

```python
model=keras.models.load_model('/content/drive/MyDrive/Project/blstm_model.h5')
```

```python
def encoding(string):
  #string="sam2an gedara yai"
  words=string.split()
  words=[list(i) for i in words]
  sentence=np.zeros((len(words),22,60))
  for i,word in enumerate(words):
    for j,letter in enumerate(word):
      vectorized=inf_english_dic.get(letter)
      if vectorized == None:
        k = 0
      else:
        k = vectorized
      sentence[i][j][k]=1
    sentence[i][j+1:][:]=1
  return sentence
```

```python
def translation(string):
  encoded=encoding(string)
  results=model.predict(encoded)
  
  numbers=np.argmax(results,axis=2)
  for i, word in enumerate(numbers):
    for j, letter in enumerate(word):
      if word[j]==word[j-1]==59:
        word[j-1:]=0
  l=''
  for i in numbers:
    for j in i:
      l+=inference_dic[j]
  sinhalized=" ".join(l.split())

  return sinhalized
```
